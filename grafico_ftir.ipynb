## **A\***

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.rcParams.update({
    "font.family": "serif",
    "mathtext.fontset": "cm",
    "font.size": 12 
})

def formatar_espaco(x, pos):
    return f"{x:,.0f}".replace(",", " ")

file_path = "dados/ftir_A_1.txt"

X_wave_number_A = []
y_tramistance_A = []

with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()

        if line.startswith("##"):
            continue
        if not line:
            continue
        
        parts = line.split()
        wave_number = float(parts[0])
        intensity = float(parts[1])
        X_wave_number_A.append(wave_number)
        y_tramistance_A.append(intensity)

X_wave_number_A = np.array(X_wave_number_A)
y_tramistance_A = np.array(y_tramistance_A)

plt.figure(figsize=(14,6), dpi=300)
plt.plot(X_wave_number_A, y_tramistance_A, color="dodgerblue", linewidth=0.9)
# plt.plot(X_wave_number_A, 100 - y_tramistance_A, color="dodgerblue", linewidth=0.9)
plt.xlabel("Nmero de onda (cm)", fontsize=12)
plt.ylabel("Transmitncia (%)", fontsize=12)
# plt.ylabel("Absorbncia (a.u.)", fontsize=12)
plt.title("Espectro FTIR", fontweight="bold", fontsize=16)
plt.xlim(min(X_wave_number_A), max(X_wave_number_A))
plt.grid(linestyle="--", linewidth=0.6, alpha=0.5)

ax = plt.gca()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(6))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(6))
ax.tick_params(axis="both", which="major", length=8, width=1)
ax.tick_params(axis="both", which="minor", length=4, width=0.7)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(formatar_espaco))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(formatar_espaco))

ax.invert_xaxis()

regioes = {
    (3800, 3040): "O-H",
    (2800, 3000): "C-H",
    (1100, 1000): "Si-O-Si (TO)",
    (1200, 1150): "Si-O-Si (LO)",
    (770, 790): "Si-O",
    (950, 970): "Si-OH"

}

for regiao, ligacao in regioes.items():
    inicio, fim = regiao
    plt.axvspan(inicio, fim, alpha=0.3, color="violet", label=ligacao)

plt.show()
```

## **Plotando os Grficos Juntos**

```python
plt.figure(figsize=(15,6), dpi=600)
plt.plot(X_wave_number_A, y_tramistance_A, color="dodgerblue", linewidth=1.1, label="Espectro A*", alpha=0.8)
plt.plot(X_wave_number_B, y_tramistance_B, color="crimson", linewidth=1.1, linestyle="-.", label="Espectro B*")
plt.xlabel("Nmero de onda (cm)", fontsize=15)
plt.ylabel("Transmitncia (%)", fontsize=15)
plt.title("Espectro FTIR - SiNPs", fontweight="bold", fontsize=16)
plt.xlim(min(X_wave_number_B), max(X_wave_number_B))
plt.grid(linestyle="--", linewidth=0.6, alpha=0.5)
plt.legend(fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

ax = plt.gca()
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(6))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(6))
ax.tick_params(axis="both", which="major", length=8, width=1)
ax.tick_params(axis="both", which="minor", length=4, width=0.7)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(formatar_espaco))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(formatar_espaco))
ax.invert_xaxis()

valor_minimo = min(min(y_tramistance_B), min(y_tramistance_A))
ax.set_ylim(bottom=valor_minimo, top=118)

regioes = {
    (3040, 3800): ["O-H", 10],
    (2800, 3000): ["C-H", 13],
    (1150, 1200): ["Si-O-Si", 8],
    (1000, 1100): ["Si-O-Si", 30],
    (950, 970): ["Si-OH", 5],
    (770, 790): ["Si-O", 8],
}

def encontrar_vale(regiao_x, regiao_y):
    idx = np.argmin(regiao_y)
    return regiao_x[idx], regiao_y[idx]

for regiao, [ligacao, subida] in regioes.items():
    mask = (X_wave_number_B > regiao[0]) & (X_wave_number_B < regiao[1])

    x_min_A, y_min_A = encontrar_vale(X_wave_number_A[mask], y_tramistance_A[mask])
    _, y_B_corr = encontrar_vale(X_wave_number_A[mask], y_tramistance_B[mask])

    plt.vlines(x=x_min_A, ymin=valor_minimo, ymax=y_B_corr, linewidth=0.9, linestyle="--", color="black", alpha=0.7)
    plt.text(x_min_A-5, y_B_corr +3, f"{str(round(x_min_A, 2))} cm", horizontalalignment="center", rotation=90, fontsize=8) # Nmero em cima das linhas
    plt.text(x_min_A+24, y_min_A/2 + subida, f"{ligacao}", horizontalalignment="center", rotation=90, fontsize=10, color="indigo") # Ligao

plt.legend(fontsize=13)
plt.show()
```
